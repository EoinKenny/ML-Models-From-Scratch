# Linear Regression with Gradient Descent in TensorFlow

```python
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Get suitable dataset
df = pd.read_csv("hubble_constant.csv")
```

```python
# Actual values
y = list(df.recession_velocity)
x = list(df.distance)
```

```python
# Placeholders (answers to find...)
b0 = tf.Variable(0, dtype=tf.float32, name="b0")  # Slope
b1 = tf.Variable(0, dtype=tf.float32, name="b1")  # Bias

# b0 = tf.placeholder(tf.float32, [None, 1] name="b0")  # Slope
# b1 = tf.placeholder(tf.float32, [None, 1], name="b1")  # Bias
```

```python
# Linear Regression
output = x*b0 + b1
```

```python
# Cost function
# Minimise the residules
cost = tf.reduce_mean((y-output)**2)
```

```python
train = tf.train.GradientDescentOptimizer(0.01).minimize(cost)
```

```python
init = tf.global_variables_initializer()
session = tf.Session()
session.run(init)
```

```python
weight_change = list()
bias_change = list()

for _ in range(2000):
    weight_change.append(session.run(b0))
    bias_change.append(session.run(b1))
    session.run(train)
```

```python
print(session.run(b1), session.run(b0))
```

```python
plt.scatter(x, y)
plt.plot([0, 2], [session.run(b1), 2*session.run(b0) + session.run(b1)], c="r")

# y = x*b0 + b1
```

```python
plt.plot(list(range(2000)), weight_change)
```

```python
plt.plot(list(range(2000)), bias_change)
```
